```python
import pandas as pd
```

```python
tumor = pd.read_csv("data/tumorgrowth_long.csv")
```

```python
tumor
```

```python
tumor_tidy = tumor.melt(id_vars=["Group", "Grp", "ID"], var_name="day", value_name="size")
```

```python
tumor_tidy
```

```python
avg_mean = tumor_tidy.groupby(["Group", "day"]).agg(avg = ("size", "mean")).reset_index()
```

```python
avg_mean.info()
```

```python
avg_mean = (tumor_tidy
            .groupby(["Group", "day"])
            .agg(avg = ("size", "mean"))
            .reset_index()
            .assign(day=pd.to_numeric(tumor_tidy["day"]))
            .sort_values(["Group", "day"])
)
```

```python
avg_mean.info()
```
